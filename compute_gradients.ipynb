```python
%matplotlib qt
from pathlib import Path
import numpy as np
import pandas as pd
from brainspace.gradient import GradientMaps
import seaborn as sns
import matplotlib.pyplot as plt
from nilearn.plotting import plot_matrix
```

```python
subjects_mrtrix_dir = "/Volumes/G_USZ_ORL$/Research/ANTINOMICS/payam/subjects_mrtrix_dir"
subject_id = "bctw"
atlas = "glasser"
conn_fname = Path(subjects_mrtrix_dir) / subject_id / "conn" / f"{atlas}_conn.csv"
conn = pd.read_csv(conn_fname, header=None).to_numpy()
np.fill_diagonal(conn, 1)
```

```python
labels_fname = "/usr/local/mrtrix3/share/mrtrix3/labelconvert/hcpmmp1_ordered.txt"
labels = np.loadtxt(labels_fname, dtype=str)[:, 1]

fig, ax = plt.subplots(figsize=(10, 8))
plot_matrix(conn, labels=labels, axes=ax, colorbar=True, vmax=0.9, vmin=0)
```

```python
gm = GradientMaps(
                        n_components=3,
                        approach="dm",
                        kernel=None,
                        alignment='procrustes',
                        random_state=0
                        )
# gm.fit(reference)
gm.fit(conn)
g1, g2, g3 = gm.gradients_.T
```
